<h2> Extracting title and text from the urls and storing the content in their respective url_id text files.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# load the URLs and URL IDs from the Excel file
df = pd.read_excel('Input.xlsx')
urls = df['URL'].tolist()
url_ids = df['URL_ID'].tolist()

# loop through each URL and extract the article title and text
for i, url in enumerate(urls):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # find the article title
    title_elem = soup.find('h1')
    if title_elem is None:
        print(f"No title found for URL ID {url_ids[i]}")
        title = ''
    else:
        title = title_elem.get_text()

    # find the article text
    article = soup.find('article')
    if article is None:
        print(f"No article found for URL ID {url_ids[i]}")
        text = ''
    else:
        # exclude footer and img tags
        for tag in article.select('footer,img'):
            tag.decompose()

        # extract text from remaining tags
        text = ''
        for p in article.find_all(['p', 'ol', 'ul'], recursive=True):
            text += p.get_text(strip=True) + '\n'

    # save the title and text to a file with the URL ID as its name
    url_id = url_ids[i]
    with open(f"{'url_'+str(url_id)}.txt", 'w', encoding='utf-8') as f:
        f.write(title + '\n\n' + text)
```

<h2> Extracting text from urls and calculating 'Positive Score', 'Negative Score', 'Polarity Score', 'Subjectivity Score'

```python

# load the URLs and URL IDs from the Excel file
df = pd.read_excel('Input.xlsx')
urls = df['URL'].tolist()
url_ids = df['URL_ID'].tolist()

# load the positive and negative dictionaries
with open('positive-words.txt', 'r', encoding='utf-8') as f:
    positive = set(f.read().splitlines())

with open('negative-words.txt', 'r', encoding='latin-1') as f:
    negative = set(f.read().splitlines())

# create a list to store the scores for each URL
scores = []

# loop through each URL and extract the article title and text
for i, url in enumerate(urls):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # find the article title
    title_elem = soup.find('h1')
    if title_elem is None:
        print(f"No title found for URL ID {url_ids[i]}")
        title = ''
    else:
        title = title_elem.get_text()

    # find the article text
    article = soup.find('article')
    if article is None:
        print(f"No article found for URL ID {url_ids[i]}")
        text = ''
    else:
        # exclude footer and img tags
        for tag in article.select('footer,img'):
            tag.decompose()

        # extract text from remaining tags
        text = ''
        for p in article.find_all(['p', 'ol', 'ul'], recursive=True):
            text += p.get_text(strip=True) + ' '

        # tokenize the text using NLTK
        tokens = nltk.word_tokenize(text)

        # calculate the scores
        pos_score = sum([1 for word in tokens if word in positive])
        neg_score = sum([1 for word in tokens if word in negative]) * -1
        polarity_score = min(max((pos_score - neg_score) / (pos_score + neg_score + 0.000001), -1), 1)
        subjectivity_score = (pos_score + neg_score) / ((len(tokens) + 0.000001))

        # add the scores to the list
        scores.append([url_ids[i], pos_score, neg_score, polarity_score, subjectivity_score])

# write the scores to a CSV file
with open('scores.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['URL_ID', 'Positive Score', 'Negative Score', 'Polarity Score', 'Subjectivity Score'])
    writer.writerows(scores)
```

<h2> Extracting text from urls and calculating 'Average Sentence Length', 'Percentage of Complex Words', 'Fog Index'

```python
import csv
import pandas as pd
import requests
from bs4 import BeautifulSoup
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt') # download the necessary NLTK tokenizer data

# load the URLs and URL IDs from the Excel file
df = pd.read_excel('Input.xlsx')
urls = df['URL'].tolist()
url_ids = df['URL_ID'].tolist()

# load the positive and negative dictionaries
with open('positive-words.txt', 'r', encoding='utf-8') as f:
    positive = set(f.read().splitlines())

with open('negative-words.txt', 'r', encoding='latin-1') as f:
    negative = set(f.read().splitlines())

# create a list to store the scores for each URL
Reliability_score = []

# loop through each URL and extract the article title and text
for i, url in enumerate(urls):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # find the article title
    title_elem = soup.find('h1')
    if title_elem is None:
        print(f"No title found for URL ID {url_ids[i]}")
        title = ''
    else:
        title = title_elem.get_text()

    # find the article text
    article = soup.find('article')
    if article is None:
        print(f"No article found for URL ID {url_ids[i]}")
        text = ''
    else:
        # exclude footer and img tags
        for tag in article.select('footer,img'):
            tag.decompose()

        # extract text from remaining tags
        text = ''
        for p in article.find_all(['p', 'ol', 'ul'], recursive=True):
            text += p.get_text(strip=True) + ' '

        # tokenize the text using NLTK
        tokens = nltk.word_tokenize(text)

       

       
        # calculate the Gunning Fox index
        sentences = nltk.sent_tokenize(text)
        num_sentences = len(sentences)
        num_words = len(tokens)
        num_complex_words = len([word for word in tokens if len(word) > 2 and word.isalpha() and len(nltk.pos_tag([word])[0][1]) > 1])
        avg_sentence_length = num_words / num_sentences
        percent_complex_words = num_complex_words / num_words
        fog_index = 0.4 * (avg_sentence_length + percent_complex_words)

        # add the scores and Gunning Fox index to the list
        Reliability_score.append([url_ids[i], avg_sentence_length, percent_complex_words, fog_index])

# write the scores to a CSV file
with open('Reliability_score.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['URL_ID', 'Average Sentence Length', 'Percentage of Complex Words', 'Fog Index'])
    writer.writerows(Reliability_score)
```

<h2> Extracting text from urls and calculating 'Average Number of Words per Sentence', 'Complex Word Count', 'Word Count'

```python


# load the URLs and URL IDs from the Excel file
df = pd.read_excel('Input.xlsx')
urls = df['URL'].tolist()
url_ids = df['URL_ID'].tolist()

# create a list to store the scores for each URL
word_count = []

# load the CMU Pronouncing Dictionary
cmudict = nltk.corpus.cmudict.dict()

# loop through each URL and extract the article title and text
for i, url in enumerate(urls):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # find the article title
    title_elem = soup.find('h1')
    if title_elem is None:
        print(f"No title found for URL ID {url_ids[i]}")
        title = ''
    else:
        title = title_elem.get_text()

    # find the article text
    article = soup.find('article')
    if article is None:
        print(f"No article found for URL ID {url_ids[i]}")
        text = ''
    else:
        # exclude footer and img tags
        for tag in article.select('footer,img'):
            tag.decompose()

        # extract text from remaining tags
        text = ''
        for p in article.find_all(['p', 'ol', 'ul'], recursive=True):
            text += p.get_text(strip=True) + ' '

        # tokenize the text using NLTK
        tokens = nltk.word_tokenize(text)

        # remove stop words and punctuation from tokens
        words = [word.lower() for word in tokens if word.lower() not in stop_words and word.isalpha()]

        # calculate the average number of words per sentence
        sentences = nltk.sent_tokenize(text)
        num_sentences = len(sentences)
        num_words = len(words)
        avg_num_words_per_sentence = num_words / num_sentences

        # calculate the complex word count
        complex_words = []
        for word in words:
            if word in cmudict and len(cmudict[word][0]) > 2:
                complex_words.append(word)
        num_complex_words = len(complex_words)

        # calculate the word count
        num_words = len(words)

        # add the scores to the list
        word_count.append([url_ids[i], avg_num_words_per_sentence, num_complex_words, num_words])

# write the scores to a CSV file
with open('word_count.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['URL_ID', 'Average Number of Words per Sentence', 'Complex Word Count', 'Word Count'])
    writer.writerows(word_count)
```

<h2> Extracting text from urls and calculating 'Syllable Count' ,'Personal Pronoun Count', 'Average Word Length'

```python
from nltk.corpus import cmudict, stopwords
# load the URLs and URL IDs from the Excel file
df = pd.read_excel('Input.xlsx')
urls = df['URL'].tolist()
url_ids = df['URL_ID'].tolist()

# create a list to store the scores for each URL
syllable_word_count = []

# load the CMU Pronouncing Dictionary
cmudict_dict = cmudict.dict()

# define regex pattern for personal pronouns
pronoun_pattern = re.compile(r'\b(I|we|my|mine|our|ours|us)\b', re.IGNORECASE)

# loop through each URL and extract the article title and text
for i, url in enumerate(urls):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # find the article title
    title_elem = soup.find('h1')
    if title_elem is None:
        print(f"No title found for URL ID {url_ids[i]}")
        title = ''
    else:
        title = title_elem.get_text()

    # find the article text
    article = soup.find('article')
    if article is None:
        print(f"No article found for URL ID {url_ids[i]}")
        text = ''
    else:
        # exclude footer and img tags
        for tag in article.select('footer,img'):
            tag.decompose()

        # extract text from remaining tags
        text = ''
        for p in article.find_all(['p', 'ol', 'ul'], recursive=True):
            text += p.get_text(strip=True) + ' '

        # tokenize the text using NLTK
        tokens = nltk.word_tokenize(text)

        # remove stop words and punctuation from tokens, and count syllables per word
        syllable_count = 0
        total_word_length = 0
        num_words = 0
        for word in tokens:
            if word.lower() not in stopwords.words('english') and word.isalpha():
                # count number of syllables in word
                if word.endswith('es') or word.endswith('ed'):
                    syllable_count += max(1, sum([1 for c in word if c.lower() in 'aeiou']))
                else:
                    syllable_count += max(1, len(re.findall(r'[aeiouy]+', word.lower())))

                # calculate total word length
                total_word_length += len(word)
                num_words += 1

        # calculate the personal pronoun count
        personal_pronouns = len(re.findall(pronoun_pattern, text))

        # calculate the average word length
        avg_word_length = total_word_length / num_words if num_words > 0 else 0

        # add the scores to the list
        syllable_word_count.append([url_ids[i], syllable_count,personal_pronouns, avg_word_length])

# write the scores to a CSV file
# write the scores to a CSV file
with open('syllable_word_count.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['URL_ID', 'Syllable Count' ,'Personal Pronoun Count', 'Average Word Length'])
    writer.writerows(syllable_word_count)

```
